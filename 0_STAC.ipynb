## STAC: SpatioTemporal Asset Catalog

STAC (SpatioTemporal Asset Catalog) is an open specification that provides a common language for describing geospatial information. It defines a standard way to expose collections of spatial and temporal data so that they can be easily indexed, discovered, and queried. STAC is widely adopted in the cloud-native geospatial community and underpins many large-scale data catalogs such as Microsoft Planetary Computer and NASA's Earthdata.

A STAC catalog organizes data into **Items** (individual assets, e.g. a single scene or time step), **Collections** (grouped Items sharing common metadata), and **Catalogs** (top-level containers linking to Collections or other Catalogs). Assets can point to any cloud-native format — COGs, Zarr stores, NetCDF files, and more.
