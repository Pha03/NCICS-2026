# Check: does stac-geoparquet preserve `storage:schemes`?

Read the geoparquet file with pandas and inspect whether `storage:schemes` is retained as a column.

```python
import pandas as pd
import json

PARQUET = "stac_out/dynamical.parquet"

df = pd.read_parquet(PARQUET)
print(f"Shape: {df.shape}")
print(f"\nColumns:\n{list(df.columns)}")
```

```python
print("'storage:schemes' column present:", "storage:schemes" in df.columns)

if "storage:schemes" in df.columns:
    print("\nValues:")
    for i, val in enumerate(df["storage:schemes"]):
        print(f"  row {i}:", val)
else:
    print("\nAll columns containing 'storage':")
    print([c for c in df.columns if "storage" in c.lower()])
    print("\nAll columns containing 'scheme':")
    print([c for c in df.columns if "scheme" in c.lower()])
```

```python

# Inspect the assets column for the first item
if "assets" in df.columns:
    raw = df["assets"].iloc[0]
    assets = raw if isinstance(raw, dict) else json.loads(raw)
    print("Assets for first item:")
    # Convert any non-serializable types for display
    def to_serializable(obj):
        if isinstance(obj, dict):
            return {k: to_serializable(v) for k, v in obj.items()}
        if isinstance(obj, (list, tuple)):
            return [to_serializable(v) for v in obj]
        try:
            json.dumps(obj)
            return obj
        except TypeError:
            return str(obj)
    print(json.dumps(to_serializable(assets), indent=2))
```
